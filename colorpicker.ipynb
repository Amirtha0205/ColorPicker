# Get the color for the respective RGB value

```python
def getColorName(R,G,B):
    global name
    minimum = 10000
    for i in range(len(d)):
        d0 = abs(R- int(d.loc[i,"R"])) + abs(G- int(d.loc[i,"G"]))+ abs(B- int(d.loc[i,"B"]))
        if(d0<=minimum):
            minimum = d0
            name = d.loc[i,"Name"]
    return name
```
